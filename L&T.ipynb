```python
# Declaring our independent features(X1) and dependent feature(Y1) in train1 file
X1 = train1[['segment','date','month','year','weekday','remarks']]
Y1 = train1[['case_count']]
# Declaring our independent features(X1) and dependent feature(Y1) in train2 file
X2 = train2[['segment','date','month','year','weekday','remarks']]
Y2 = train2[['case_count']]
```

```python
# Importing Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
# Fitting the model for segment 1
reg1 = reg.fit(X1,Y1)
# Predicting case_count for segment 1 
Y_pred1=reg1.predict(test1)
```

```python
# Fitting the model for segment 2
reg2 = reg.fit(X2,Y2)
# Predicting case_count for segment 2
Y_pred2=reg2.predict(test2)
```

```python
# Print regression coefficicients and regression intercept for segment 2
print(reg2.coef_) 
print(reg2.intercept_)
```

```python
# Adding our predicted case counts of both segments in their respective dataframe.
df1['case_count']= Y_pred1
df2['case_count']= Y_pred2
```
